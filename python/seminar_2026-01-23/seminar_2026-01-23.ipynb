### Data structures in Python

```python
# Text Type: str
a = "Hello World"
print(type(a))
# Numeric Types: int, float, complex
b = 20
print(type(b))
c = 20.5
print(type(c))
d = 20 + 100j
print(type(d))
# Sequence Types: list, tuple, range
e = ["apple", "banana", "cherry"]
print(type(e))
f = ("apple", "banana", "cherry")
print(type(f))
g = range(6) 
print(type(g))
# Mapping Type: dict
h = {
    "name" : "John", 
    "age" : 36
    }
print(type(h))
# Set Types: set, frozenset
i = {"apple", "banana", "cherry"}
print(type(i))
k = frozenset({"apple", "banana", "cherry"})
print(type(k))
# Boolean Type:   bool
g = True
print(type(g))
# Binary Types:   bytes, bytearray, memoryview
m = b"Hello"
print(type(m))
e = bytearray(5)
print(type(e))
z = memoryview(bytes(5))
print(type(z))
# None Type:      NoneType
y = None
print(type(y))
```
