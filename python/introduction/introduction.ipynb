# Data types

```python
# Text Type: str
a = "Hello World"
print("Type of a:", type(a))



# Numeric Types: int, float, complex
b = 20
print("Type of b:", type(b))

c = 20.5
print("Type of c:", type(c))
d = 20 + 100j
print("Type of d:", type(d))



# Sequence Types: list, tuple, range
e = ["apple", "banana", "cherry"]
print("Type of e:", type(e))

f = ("apple", "banana", "cherry")
print("Type of f:", type(f))
g = range(6) 
print("Type of g:", type(g))



# Mapping Type: dict
h = {
    "name" : "John", 
    "age" : 36
    }
print("Type of h:", type(h))


# Set Types: set, frozenset
i = {"apple", "banana", "cherry"}
print("Type of i:", type(i))

k = frozenset({"apple", "banana", "cherry"})
print("Type of k:", type(k))



# Boolean Type:   bool
g = True
print("Type of g:", type(g))


# Binary Types:   bytes, bytearray, memoryview
m = b"Hello"
print("Type of m:", type(m))

e = bytearray(5)
print("Type of e:", type(e))

z = memoryview(bytes(5))
print("Type of z:", type(z))



# None Type:      NoneType
y = None
print("Type of y:", type(y))
```

# Data structures in Python

## List/Array

### Initialization/Declaration

```python
a = [] 
b = ["DNA", "RNA", "Protein"]
c = [1, 2, 3, 4, 5]
d = [1, "Hello", 3.4, True]

print("a:", a)
print("b:", b)
print("c:", c)
print("d:", d)
```

### Methods

#### Access element(s) method

```python
a = [0, 1, 2, 3 , 4, 5]
b = ["DNA", "RNA", "Protein"]
c = ["a", "b", "c", "d", "e"]

# Access element(s) method
print("Access element(s) method")
print("\n") # newline character
print("a[2]:", a[2])
print("b[-1]:", b[-1])
print("c[0]:", c[0])

print("c[1:3]:", c[1:3]) # tu 1 den < 3
print("c[2:4]:", c[2:4]) # tu 2 den < 4

print("c[2:]:", c[2:]) # tu 2 den < ~
print("c[:2]:", c[:2]) # tu ~ den < 2
print("c[-4:-1]:", c[-4:-1]) # tu -4 den < -1
```

#### Append method

```python
a = [0, 1, 2, 3 , 4, 5]
b = ["DNA", "RNA", "Protein"]
c = ["a", "b", "c", "d", "e"]

# Append method
print("Append method")
print("\n") # newline character

a.append(5)
print("a.append(5):", a)

a.append("hello world")
print("a.append('hello world'):", a)
```

#### Join method

```python
a = [0, 1, 2, 3 , 4, 5]
b = ["DNA", "RNA", "Protein"]
c = ["a", "b", "c", "d", "e"]

# Join method
print("Join method")
print("\n") # newline character

print("a + c:", a + c)
print("c + a:", c + a)
print("a + b + c:", a + b + c)
```

#### Pop method

```python
a = [0, 1, 2, 3 , 4, 5]
b = ["DNA", "RNA", "Protein"]
c = ["a", "b", "c", "d", "e"]

## Pop methods
print("Pop methods")
print("\n") # newline character

x = a.pop(1)
print("x:", x)
print("a:", a)

y = b.pop() 
print("y:", y)
print("b:", b)
```

# Controll flow statements

## Loop statements

### For

```python
a = 1
print("a before:", a)
for i in range(10): # i from 0 to 9
    print("i:", i)
    a = a + 1
print("a after:", a)

print("\n") # newline character

for i in range(5): # i from 0 to 4
    print("i:", i)
    print("hello world")
```

### While

```python
a = 0
print("a before:", a)
while a < 4:
    a = a + 1
    print("hello world")
print("a after:", a)
```
